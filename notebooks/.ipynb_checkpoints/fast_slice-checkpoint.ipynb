```python


def register(section: Section, atlas):
    width, height = section.image.width,section.image.height
    image_coords = np.mgrid[:width, :height]
    image_coords_h = np.zeros((4, width, height))
    image_coords_h[:2] += image_coords
    image_coords_h[3] += 1
    inds = image_coords_h.reshape(4, -1)
    coords = section.affine_transform @ inds
    atlas_coords = coords[:3, :] / atlas.resolution_um
    atlas_coords_i = np.round(atlas_coords).astype(int)
    x, y, z = atlas_coords_i
    vals = atlas.volume[x, y, z]
    registered_slice = vals.reshape(width, height)

    return registered_slice

s = section.translate(dz=4000).rotate(dx=70)
%timeit atlas_slice = register(s, atlas)
# plt.imshow(atlas_slice);
```
